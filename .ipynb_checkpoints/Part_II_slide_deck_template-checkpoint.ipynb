## Investigation Overview

This investigation was performed on the January 2022 data of the National Bureau of Transportation Statistics that gathers information about the lateness in flights by US carriers.

We performed and initial exploration of the data, performed some gathering and wrangling and will be focusing on the flights that arrived late (delay > 1 minute). We'll be checking the main reasons behind the delays and the correlation between some of the factors involved.

### Key Insights


* Airlines are the main factor that explain delays in the US. 
* 95\% of the arrival delay is explained by the departure delay, the other 5\% is mainly explained by National Air Service related delays in arrival procedures.
* Hubs are the main pain point for Airlines.
* Bigger airlines tend to have a more skewed distribution towards greater delays than smaller airlines.


## Dataset Overview

The Dataset contains data of all flights performed by US Carriers in the month of January 2022. It specifies the delay on departure, on arrival, the distance and time in the air, as well as a break down on the reason for the delay. 

This breakwdown has 5 components: 
* Carrier Delay: Delays caused by problems in the airline's operation.
* National Air System Delay: Delays related to the management of airspace and ground operations.
* Weather delay: Delays related to weather phenomena.
* Security Delay: Delays caused by security, TSA and other agencies.
* Late Aircraft Delay: Delays caused by the aircraft intended arriving late from the previous flight.



```python
# import all packages and set plots to be embedded inline
import datetime, warnings, scipy 
import pandas as pd
import matplotlib as mpl
import folium
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import requests
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('flight_delays.csv')
# Load the Airport and airlines database.
df_airports = pd.read_csv('airports.csv')
df_carriers = pd.read_csv('carriers.csv')

# Load the final version of the Dataframe with the wrangling steps performed:
df_new = pd.read_csv('final_df.csv')
```

## How is Traffic Distributed in the US?

> We can see a clear distribution of traffic in the US on the following map.
* Traffic is heavily distributed in both East and West Coast.
* The main Airports by traffic are
    * Dallas
    * Denver
    * Chicago
    * Atlanta
* This airports are hubs for the main US carriers. They use a Hub and Spoke model that redirects regional traffic through main hubs, where flights to the main cities in the US are operated.

> **Rubric Tip**: All plots in the slideshow are polished, meaning all plots have a title, labeled x/y axes (with units), x/y ticks, and legends. 

```python
# We need to create a list of the airports and the number of flights
value = pd.DataFrame(df['ORIGIN'].value_counts())
value.reset_index(inplace = True)

# Join with Airport data
traffic_data = df_airports.merge(value, left_on= 'iata', right_on = 'index',how = 'left')

# Drop airports without data
traffic_data = traffic_data.dropna(axis = 0, subset=['ORIGIN'])

# Create an encoding for traffic
traffic_data['n_flights'] = np.select(
    [traffic_data['ORIGIN'].between(0,1000, inclusive = False),
    traffic_data['ORIGIN'].between(1000,10000, inclusive = False),
    traffic_data['ORIGIN'].between(10000,20000, inclusive = False),
    traffic_data['ORIGIN'].between(20000,30000, inclusive = False)],
    [
        1,
        2,
        3,
        4
    ],
    default = 5
    )
traffic_data['color'] = np.select(
    [traffic_data['ORIGIN'].between(0,1000, inclusive = False),
    traffic_data['ORIGIN'].between(1000,10000, inclusive = False),
    traffic_data['ORIGIN'].between(10000,20000, inclusive = False),
    traffic_data['ORIGIN'].between(20000,30000, inclusive = False)],
    [
        'green',
        'yellow',
        'orange',
        'red'
    ],
    default = 'black'
    )

# Vamos a utilizar la librería folium que nos permite crear mapas interactivos para formatos HTML.
# Creamos el mapa
us_airports = folium.Map(location = [37.09024, -95.712891], zoom_start = 2.5)

traffic_data.apply(lambda row:folium.CircleMarker(location = [row["lat"], row["long"]], radius = row['n_flights'], color = row['color']).add_to(us_airports), axis = 1);
us_airports
```

## How Are Delays distributed?

> We applied a logarithmic transformation to the data, since it was extremely right skewed. 
* We can observe a bi-modal distribution with two spikes around:
    * 15 minutes
    * 30-45 Minutes

* The graph also shows how the vast majority of delays are between 15 and 100 minutes.

```python
# Create our plot
#fig, ax = plt.subplots()
# Set Logarithmic scale
sb.set(style='darkgrid')
fig = sb.displot(data = df_new, x = 'ARR_DELAY_NEW', log_scale= True, color='blue', height=6, aspect=1.5);
# Adapt ticks for new scale
xticks = [15, 20, 25, 30, 45, 60,100,500]
plt.xticks(xticks, xticks);
fig.set_axis_labels('Minutes of lateness', '');
#ax.set_ylabel(None);
```

## Who causes the greater delays?

* Carriers are responsible for an average 30 minutes of delay directly caused by their operations.

* Late aircrafts are the second leading factor affecting delays.

    * Delays are mostly a chain reaction, and late aircraft can have a huge impact on the process. All the Airline processes are handled in a timely fashion. Every minute of delay puts a strain on the structure leading to more delays.
    
    * All of the other factors can feedback into a plane becoming late. In this case, we can think of the other categories leading into an original delay, that will become a Late Aircraft Delay for the rest of the flights that an aircraft performs in a day.

* Weather and Security produced less than 5 minutes in average delays.

* National Air System caused an average 10 minutes delay.

```python
# Set the Data
data = df[['LATE_AIRCRAFT_DELAY', 'SECURITY_DELAY', 'NAS_DELAY', 'WEATHER_DELAY', 'CARRIER_DELAY']].mean()
mean = pd.DataFrame(data, columns= ['mean_delay'])

#Plot
fig, ax = plt.subplots()
labels = ['Carrier', 'Late Aircraft', ' National Air System', 'Weather', 'Security']
sb.barplot(data = mean, x = data.index,  y = 'mean_delay', color = sb.color_palette()[0],
                        order= ['CARRIER_DELAY', 'LATE_AIRCRAFT_DELAY', 'NAS_DELAY', 'WEATHER_DELAY', 'SECURITY_DELAY']);
plt.xticks(rotation = 25);
ax.set_xticklabels(labels);
plt.title('Average delay by Category');
plt.ylabel('Mean delay in Minutes');
```

## Which airlines have a worst on-time ratio?

* Smaller airlines tend to have a more balanced equilibrium between groups (for example American Eagle and Midwest Airlines have an almost equal balance).
* Bigger airlines such as Delta, Skywest, Southwest have a distribution more skewed to bigger delays. 

* From the last two graphs we can see, that the bigger the operative and concentration of operations, the bigger the delays.

```python
#delay = lambda x:((0,1)[x > 10], 2)[x > 45]
#df_new['delay_group'] = df_new['DEP_DELAY_NEW'].apply(delay)

# Countplot 
fig = plt.figure(1, figsize = (15,15))
ax = sb.countplot(data = df_new, y = 'Airline', hue = 'delay_group')
ax.yaxis.label.set_visible(False)

# Set up a legend and show the plot
legend = plt.legend()
legend.get_texts()[0].set_text('On Time (Delay < 10 min)')
legend.get_texts()[1].set_text('Short Delay (10 < Delay < 45 min)')
legend.get_texts()[2].set_text('Big Delay (Delay > 45 min)')
plt.title('Number of flights per Delay Group', fontweight="bold")
plt.xlabel('Number of Flights Per Delay Group', fontweight="bold")
plt.show()
```

## Is there a relationship between short, medium and long haul flights and lateness?

We divided the flights in terms of Air Time: 
* Short Haul: 0 to 180 minutes.
* Medium Haul: 180 to 360 minutes.
* Long Haul: More than 360 minutes.

Key takeaways:

* The longer the flight, the shorter the average delay.

* Short haul flights have the more severe distribution of delays.

    * This confirms the data on the second slide, where we defined the cascading effects that delays can have.

```python
# Plot
markers = [['Short Haul', 'o'], ['Medium Haul', 's'], ['Long Haul', 'x']]


plt.figure(figsize = (12,8))
for cat, marker  in markers:
    df_cat = df_new[df_new['air_time_group'] == cat]
    plt.scatter(data = df_cat, x = 'DEP_DELAY_NEW', y = 'ARR_DELAY_NEW', marker = marker, alpha= 0.4);

plt.legend(['Short Haul', 'Medium Haul', 'Long Haul']);
plt.xlabel('Departure Delay in Minutes')
plt.ylabel('Arrival Delay in Minutes')
plt.title('Behaviour of Short, Medium and Long Haul Flights', fontweight="bold");
```

#### How are the types of delays related between them?

* We can see a 95% correlation between Arrival Delay and Departure Delay.

* Carrier and Late Aircraft delay are the most correlated with Departure Delay. 

* Air Service Delay is positively correlated with Arrival Delays and negatively correlated with departure delays:

    * This exemplifies the complexity and frailness of arrival procedures.


* Security delay is not correlated and seems to not have affected air traffic in the month of January in the USA.

```python
# select just the delays
df_corr = df_new[['DEP_DELAY_NEW', 'ARR_DELAY_NEW', 'CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']]
labels = ['Departure Delay', 'Arrival Delay', 'Carrier Delay', 'Weather Delay', 'Air Service Delay', 'Security Delay', 'Late Aircraft Delay']
# Plot the heatmap of correlations
plt.figure(figsize = (12,8))
sb.heatmap(df_corr.corr(), annot = True, fmt = '.2f', cmap = 'vlag_r', center = 0, xticklabels=labels, yticklabels=labels);
plt.title('Correlation between different Delays', fontweight="bold")
```
